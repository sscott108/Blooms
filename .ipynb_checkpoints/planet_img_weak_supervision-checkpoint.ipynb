```python
#imports
import rasterio
from rasterio.windows import Window
from pyproj import Transformer
import numpy as np
import os
from os import listdir
from os.path import isfile, join
np.random.seed(42)
import pandas as pd
import copy
import matplotlib
import matplotlib.pyplot as plt
from collections import Counter

import math
import pyproj
import pickle as pkl
import geojson
import cv2
from PIL import Image, ImageEnhance
import cv2
import warnings
warnings.filterwarnings('ignore')
os.getcwd()
```

```python
from collections import defaultdict
image_files = defaultdict(list)
label_files = defaultdict(list)

for image_file in sorted(listdir('../spotone/crops')):
    date = image_file.split('_')[0]  
    image_files[date].append(image_file)

for label_file in sorted(listdir('lbls_tifs')):
    year = label_file.split('.')[1][:4]
    date = label_file.split('.')[2]
    parts = year+date 
    label_files[parts].append(label_file)

# Match images and labels by date, ensuring one per day
matches = []
for date in image_files:
    if date in label_files:
        matches.append((image_files[date][0], label_files[date][0]))  # Only pick the first instance for each day

# # Print or return the matches
# for img, lbl in matches:
#     print(f"Image: {img}, Label: {lbl}")
```

```python
markers = set()

for file in sorted(listdir('downloads')):
    parts = file.split('.')[1]
    markers.add(parts)
```

```python
markers
```

```python
labels = set()
for folder in sorted(os.listdir('all_data')):
    if folder not in [".DS_Store", ".ipynb_checkpoints"]:
        # Iterate over the files
        folder_path = os.path.join('all_data', folder)
#         print(folder_path)
        for file in sorted(os.listdir(os.path.join('all_data', folder))):
            if file.endswith('.tif'):
                parts = file.split('.')
                name = file.split('.')[-3]
                marker = parts[1]
#                 print(parts, name, marker)
                if marker in markers and name == 'CIcyano':
                    if marker not in labels:
                        labels.add(os.path.join(folder_path, file))
```

```python
lat, lon = 36.237286, -76.697184
crop_size = 3
lbldir = 'lbls'

for labl in labels:
    filename = labl.split('/')[-1]
    print(labl)
    with rasterio.open(labl) as rds:
        transformer = Transformer.from_crs("EPSG:4326", rds.crs, always_xy=True)
        xx, yy = transformer.transform(lon, lat)
        row, col = rds.index(xx, yy)
#         print(rds.crs)
#         # Define the crop window
        row_start = max(row - crop_size, 0)
        row_end = min(row + crop_size, rds.height)
        col_start = max(col - crop_size, 0)
        col_end = min(col + crop_size, rds.width)

        try:
            window = Window.from_slices((row_start, row_end), (col_start, col_end))
            data = rds.read(window=window)
            data = np.moveaxis(data, 0, 2)  # Move the channel axis to the last position
            img_arr = np.array(data)

            plt.imshow(img_arr)
            plt.title(f"Cropped Image around lat: {lat}, lon: {lon}")
            plt.axis('off')
            plt.show()
#             print(img_arr)

            # Uncomment to save the output image
#             output_filename = os.path.join(lbldir, filename) #.replace('.tif', '.jpg'))
#             im = Image.fromarray(np.squeeze(img_arr))  # Squeeze if it's single-channel
#             im.save(output_filename)

        except Exception as e:
            print(f"Error processing {filename}: {e}")
            pass
```
