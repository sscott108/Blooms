I recommened making a copy of the final `crop` folder and deleting the large raster images as they take up a lot of space. The `black_space` call filters out images that are either over saturated with clouds or do not have a complete capture of our area of interest, as these will not help with our analysis.
     - `main_directory` is the full path to the images downloaded. 
     - `cropped_img_folder_name` is the name of the folder in which you will store your cropped images.
     - `lat` is the latitude of your area of interest.
     - `lon` is the longitude of your area of interest.
